#Anomaly detection: Statistical methods

```python
#Copy of the dataframe to work with IQR
df_IQR = df.copy()
```

```python
#Creation of new columns to identify outliers for each column
for column in df.columns:
    # Calculate the first quartile (Q1)
    Q1 = df[column].quantile(0.25)
    # Calculate the third quartile (Q3)
    Q3 = df[column].quantile(0.75)
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Define the lower bound for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    # Define the upper bound for outlier detection
    upper_bound = Q3 + 1.5 * IQR

    # Create a new column to mark outliers for the current column
    # An outlier is marked as True if it's below the lower bound or above the upper bound, otherwise False
    df_IQR[f'{column}_flagged'] = ((df[column] < lower_bound) | (df[column] > upper_bound)).astype(int)

df_IQR.head()
```

```python
#Dataframe for only outliers
df_outliers = df_IQR.loc[:, df_IQR.columns.str.endswith('_flagged')]
df_outliers.head()
```

```python
#Column subset
subest_column = df_outliers.iloc[:, 0:6]
```

```python
#Create different columns for number of feature with outliers
df_outliers['Single_flagged'] = (subest_column.sum(axis=1) >= 1).astype(int)
df_outliers['Double_flagged'] = (subest_column.sum(axis=1) >= 2).astype(int)
df_outliers['Triple_flagged'] = (subest_column.sum(axis=1) >= 3).astype(int)
df_outliers['Quadruple_flagged'] = (subest_column.sum(axis=1) >= 4).astype(int)
df_outliers['Quintuple_flagged'] = (subest_column.sum(axis=1) >= 5).astype(int)
df_outliers['All_flagged'] = (subest_column.sum(axis=1) >= 6).astype(int)

df_outliers.head()
```

```python
#Setting the anomalies columns
column_anomalies = df_outliers.iloc[:, -6:]

#calculate sum, total observation and percentage of anomalies
sum_anomalies = column_anomalies.sum()
tot_obs = len(df_outliers)
percentage_anomalies = (sum_anomalies / tot_obs) * 100

#Final summary table
IQR_summary_table = pd.DataFrame({
    'anomalies': sum_anomalies,
    'Observations': tot_obs,
    'Percentage': percentage_anomalies.round(2)
})

IQR_summary_table
```

```python
#Observation with double flagged anomalies
df_IQR['Double flagged anomalies'] = df_outliers['Double_flagged']
df_IQR[df_IQR['Double flagged anomalies'] == 1]
```

When performing anomaly detection using the IQR method, a total of 4,636 anomalies were identified across the dataset. These anomalies, representing 23.73% of the total observations, include all instances where at least one feature was flagged as anomalous.

To identify anomalies with a greater business impact, we focused on observations where at least two different features were simultaneously flagged as anomalous. This refined approach yielded 422 anomalies, accounting for 2.16% of the observations. This figure aligns more closely with typical expectations for anomalies, which are generally anticipated to fall between 1% and 5%.

Furthermore, our analysis revealed no instances of anomalies occurring simultaneously across more than four features. We also found 11 anomalies (0.06% of observations) that were flagged across exactly three features concurrently.

##One-Class SVM

```python
#Create a copy of the original data set
df_SVM = df.copy()
```

```python
#Performing scaling on the data frame
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_SVM)

#Creation of dataframe with scaled data
df_scaled = pd.DataFrame(scaled_data, columns=df_SVM.columns)
df_scaled.head()
```

```python
#Performing PCA for dimensionality reduction
pca = PCA(n_components=2)
df_pca = pca.fit_transform(scaled_data)
df_pca.shape
```

```python
#One-class SVM parameter settings based on the information from IQR anomaly detections
#Selecting the combination of nu and gamma for SVM based on the percentage of anomalies
#detected with the IQR anomaly detection methods

#Parameter setting for nu and gamma
nu_values = [0.015, 0.02, 0.03, 0.04, 0.05]
gamma_values = [0.1, 0.2, 0.5, 0.7, 1.0]

#Parameter grid setting
param_grid = {
    'nu': nu_values,
    'gamma': gamma_values
}

#Model choosen
ocsvm = OneClassSVM(kernel='rbf')

#Initialisation of variables
sugggested_ocsvm = None
sugggested_nu = None
sugggested_gamma = None
min_diff_percentage = float('inf')

#Creating an empty vector for the final results
results = []

#Reitereting the model on each combination of parameter from the parameter grid
for nu in nu_values:
    for gamma in gamma_values:
        model = OneClassSVM(kernel='rbf', nu=nu, gamma=gamma)
        model.fit(df_scaled)

        # anomalies prediction
        predictions = model.predict(df_scaled)

        # Number and percentage of outliers in the prediction
        num_outliers = np.sum(predictions == -1)
        percentage_outliers = (num_outliers / len(df_scaled)) * 100

        # Difference from 2.16% (2 fetures anomalies from IQR)
        diff_from_target = abs(percentage_outliers - 2.16)

        results.append({
            'nu': nu,
            'gamma': gamma,
            'identified_outliers_count': num_outliers,
            'identified_outliers_percentage': percentage_outliers,
            'difference_from_2.16%': diff_from_target
        })

        # Parameter settings in case of suggested combination
        if diff_from_target < min_diff_percentage:
            min_diff_percentage = diff_from_target
            sugggested_nu = nu
            sugggested_gamma = gamma
            sugggested_ocsvm = model

#Printing the results
print("\nParameter combination reseaults:")
for res in results:
    print(f"nu: {res['nu']:.4f}, gamma: {res['gamma']}, Outlier (%): {res['identified_outliers_percentage']:.2f}%, Diff da 2.16%: {res['difference_from_2.16%']:.2f}%")

print(f"\nSuggested paramteres (near to 2.16% from IQR):")
print(f"  nu: {sugggested_nu}")
print(f"  gamma: {sugggested_gamma}")
```

---
After performing different combinations of nu and gamma for the one-class SVM, the combinatin closer to the IQR anomalies percentage of 2.16% results to be, nu = 0.02 and gamma = 0.1, for 2.07% of outliers. We consider also the second nearest combination of parameters (nu=0.02, gamma= 0.2), in order to compare the differences between the two models.
